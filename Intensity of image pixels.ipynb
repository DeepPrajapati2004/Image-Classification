```python
import cv2
from matplotlib import pyplot as plt
import argparse
import imutils
```

```python
path='A:/MSI_Project/HSI_Histograms/'
img=cv2.imread(path+'gsi6.jpg')
hsi=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
channels=cv2.split(img)
colors=('b','g','r')
plt.figure()
plt.title('Color histogram')
plt.xlabel('Bins')
plt.ylabel('no. of pixels')
```

```python
for (channels,colors) in zip(channels,colors):
    hist=cv2.calcHist([channels],[0],None,[256],[0,256])
    plt.plot(hist,color=colors)
    plt.xlim(0,256)
    plt.ylim(0,650000)
```
